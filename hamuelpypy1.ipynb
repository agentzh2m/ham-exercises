```python
class WeeklyVote(object):
    def __init__(self):
        self._rank = Counter()
    
    def add_vote(self, trainee):
        self._rank[trainee] += 1
    
    def top_4(self):
        return self._rank.most_common(4)
    
    def top_10_pyramid(self):
        return self.gen_pyramid(self._rank.most_common(10))
    
    def gen_pyramid(self, lst):
        # assume the list are sorted from highest to low ranking
        tri_counter = 1
        current_counter = 0
        lst_grp = []
        current_grp = []
        for trainee in lst:
            if current_counter == tri_counter:
                lst_grp.append(current_grp)
                current_grp = []
                tri_counter += 1
                current_counter = 0
            current_grp.append(trainee)
            current_counter += 1
        if len(current_grp) > 0:
            lst_grp.append(current_grp)
        str_builder = StringIO()
        transform = list(map(lambda x: len(x), lst_grp))
        max_len_group = transform.index(max(transform))
        base_length = sum(len(t.name) + 4 for t,r in lst_grp[max_len_group]) + 1
        for grp in lst_grp:
            internal_builder = StringIO()
            for t,r in grp:
                rtstr = "{}({}) ".format(t.name, lst[0][1] - r + 1)
                internal_builder.write(rtstr)
            str_builder.write(internal_builder.getvalue().center(base_length))
            internal_builder.close()
            str_builder.write("\n")
            internal_builder = StringIO()
            for i,(t,r) in enumerate(grp):
                astr = "{}".format(t.agency)
                if lst[0][1] - r + 1 > 1 and i > 0:
                    nspace = " "*(len(grp[i-1][0].name)+ len(str(grp[i-1][1]))  - len(str(grp[i-1][1])) )
                    internal_builder.write(nspace)
                internal_builder.write(astr)
            str_builder.write(internal_builder.getvalue().center(base_length))
            internal_builder.close()
            str_builder.write("\n")
        return str_builder.getvalue()
    
    def all_pyramid(self):
        return self.gen_pyramid(self._rank.most_common()[:24])
    
    def last_4(self, ranking_threshold):
        return self._rank.most_common()[:-4-1:-1]
```

```python
html_doc = requests.get('https://en.wikipedia.org/wiki/List_of_Produce_48_contestants')
soup = BeautifulSoup(html_doc.text, 'html.parser')
```

```python
tbl = soup.tbody
current_company = ""
company_span = 0
header = ['agency', 'name', 'alt_name', 'alt_lang', 'age', 'final_rank']
with open('contestant.csv', 'w', newline='') as f:
    writer = csv.writer(f, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    writer.writerow(header)
    for tr in tbl.find_all('tr')[3:]:
        ncol = 0
        contestant = dict(alt_lang=None)
        for td in tr:
            current_col = 0
            if isinstance(td, bs4.element.NavigableString):
                continue
            else:
                #we know that col 1,2,3 are company, name, age
                if (ncol == 1 or ncol == 0) and (td.span is not None):
                    if not isinstance(td.contents[0], bs4.element.NavigableString):
                        contestant['name'] = td.a.string
                        contestant['alt_name'] = td.span.string
                        if contestant['alt_lang'] is None:
                            contestant['alt_lang'] = td.span.attrs.get('lang', None)
                    else:
                        contestant['name'] = td.contents[0][:-2]
                        contestant['alt_name'] = td.span.string
                        if contestant['alt_lang'] is None:
                            contestant['alt_lang'] = td.span.attrs.get('lang', None)
                if (ncol == 1 or ncol == 2) and (td.string is not None):
                    try:
                        contestant['age'] = int(td.string)
                    except:
                        pass
                if ncol == 16 or ncol == 17:
                    try:
                        contestant['final_rank'] = int(td.string)
                    except:
                        pass
                if ncol == 0 and company_span <= 0:
                    company_span = int(td.attrs.get('rowspan', 1))
                    if td.string is None:
                        current_company = td.a.string
                    else:
                        current_company = td.string.strip()
                ncol += 1
        company_span -= 1
        contestant['agency'] = current_company
        contestant['alt_lang'] = contestant['alt_lang'] if contestant['alt_lang'] is not None else 'jp'
        writer.writerow([contestant.get(k, 0) for k in header])

    
```
