```python
G = nx.MultiDiGraph()
for index, (stake_holders, thai_name) in df[df.business_category == ''][['stake_holders', 'thai_name']].iterrows():
    for owner in stake_holders:
        if owner['percentage'] > 10:
            G.add_edge(owner['name'], thai_name)
        
pos = nx.spring_layout(G, iterations=20)
plt.figure(figsize=(30,30))
nx.draw(G, pos, node_size=100, alpha=0.8, edge_color='b', font_size=23, with_labels=True)
plt.savefig('elec_graph.png')
```
